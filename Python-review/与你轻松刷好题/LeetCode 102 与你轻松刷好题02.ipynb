```python
# 135:
# time cost: 39 mins/ with one small error - I did not add the comparing value, instead I only incremented the elements in 
# the ans list
class Solution:
    def candy(self, ratings: List[int]) -> int:
        size = len(ratings)
        ans = [1] * size
        
        for i in range(size - 1):
            if ratings[i + 1] > ratings[i]:
                ans[i + 1] = ans[i] + 1
                
        for i in range(size - 1, 0, -1):
            if ratings[i - 1] > ratings[i] and ans[i - 1] <= ans[i]:
                ans[i - 1] = ans[i] + 1
                
        return sum(ans)
    
# time complexity: O(n)
# space complexity: O(n)
```

```python
# 767. Reorganize String
# abm, 30min
class Solution:
    def reorganizeString(self, S: str) -> str:
        # two pointers 
        i, j = 0, 1
        ans = [S[i] for i in range(len(S))]
        # print(ans)
        while i < len(S) and j < len(S):
            if ans[i] == ans[j]:
                j += 1
                # print("j position", j)
            else:
                if j - i > 1:
                    # print("change!", i, j)
                    tmp = ans[i + 1]
                    ans[i + 1] = ans[j]
                    ans[j] = tmp 
                i += 1   
                # print("i positions", i)
        # print(ans)
        
        rev_ans = [ele for ele in reversed(ans)]
        # print(rev_ans)
        
        i, j = 0, 1
        while i < len(S) and j < len(S):
            if rev_ans[i] == rev_ans[j]:
                j += 1
                # print("j position", j)
            else:
                if j - i > 1:
                    # print("change!", i, j)
                    tmp = rev_ans[i + 1]
                    rev_ans[i + 1] = rev_ans[j]
                    rev_ans[j] = tmp 
                i += 1  
        
        
        for i in range(len(S) - 1):
            if rev_ans[i] == rev_ans[i + 1]:
                return ""
        
        return ''.join(rev_ans)
```

```python
# 167. Two Sum II - Input array is sorted
# Time: less than 15min
# abm/ a little mistake at beginning
class Solution:
    def twoSum(self, numbers: List[int], target: int) -> List[int]:
        size = len(numbers)
        first, second = 0, size - 1
        
        while first<size and second>=0:
            first_num, second_num = numbers[first], numbers[second]
            if first_num + second_num > target:
                second -= 1
            elif first_num + second_num == target:
                return [first + 1, second + 1]
            else:
                first += 1
'''
Time: O(n)
space:O(1)

'''
# another shorter version
class Solution:
    def twoSum(self, numbers: List[int], target: int) -> List[int]:
        l,r = 0, len(numbers) - 1
        while numbers[l] + numbers[r] != target:
            if numbers[l] + numbers[r] > target:
                r -= 1
            else:
                l += 1
        return [l + 1, r + 1]
```

```python
#88 
# Merge sorted Array:
# Time cost 43mins
# looked at all 3 solutions, finally follow the lc101c++

class Solution:
    def merge(self, nums1: List[int], m: int, nums2: List[int], n: int) -> None:
        """
        Do not return anything, modify nums1 in-place instead.
        """
        p1, p2, pos = m-1, n-1, m+n-1
        while p1>=0 and p2>=0:
            if nums1[p1] > nums2[p2]:
                nums1[pos] = nums1[p1]
                pos -= 1
                p1 -= 1
            else:
                nums1[pos] = nums2[p2]
                p2 -= 1
                pos -= 1
                
        while p2 >= 0:
            nums1[pos] = nums2[p2]
            pos -= 1
            p2 -= 1
```
